```python
# Add the Pandas dependency
import pandas as pd
import numpy as np
```

```python
# Files to load
student_data_to_load = "Resources/students_complete.csv"

# Read the student data file and store it in a Pandas DataFrame.
student_data_df = pd.read_csv(student_data_to_load)
student_data_df
```

```python
student_data_df.loc[(student_data_df['school_name'] == "Thomas High School") 
                        & (student_data_df['grade'] == "9th"), 'reading_score'] = np.nan

student_data_df.loc[(student_data_df['school_name'] == "Thomas High School") 
                        & (student_data_df['grade'] == "9th"), 'math_score'] = np.nan

student_data_df
```

```python
# Create DataFrame containing only Thomas High School Freshmen

thomas_student = student_data_df['school_name'] == "Thomas High School"

freshmen = student_data_df['grade'] == "9th"

thomas_freshmen_df = pd.DataFrame(student_data_df[thomas_student & freshmen])


thomas_freshmen_df
```

```python
# Replace currupted grades with NaN

thomas_freshmen_df.loc[:, ['reading_score', 'math_score']] = np.nan

thomas_freshmen_df.head()
```
